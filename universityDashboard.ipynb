```python
# 1Ô∏è‚É£ Importar librer√≠as necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 2Ô∏è‚É£ Cargar el dataset
df = pd.read_csv("university_student_data.csv")
print("‚úÖ Dataset cargado correctamente")
display(df.head())
```

```python
# 3Ô∏è‚É£ Informaci√≥n general
print("\nüìò Informaci√≥n del dataset:")
df.info()

print("\nüìà Estad√≠sticas descriptivas:")
display(df.describe())
```

```python
# =============================================
# 4Ô∏è‚É£ Visualizaciones exploratorias
# =============================================

sns.set(style="whitegrid", palette="muted")

# --- Retention Rate Trends Over Time ---
plt.figure(figsize=(8,5))
sns.lineplot(data=df, x="Year", y="Retention Rate (%)", marker="o")
plt.title("Retention Rate Trends Over Time")
plt.xlabel("Year")
plt.ylabel("Retention Rate (%)")
plt.show()

# --- Student Satisfaction by Year ---
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="Year", y="Student Satisfaction (%)", ci=None)
plt.title("Student Satisfaction by Year")
plt.xlabel("Year")
plt.ylabel("Satisfaction (%)")
plt.show()

# --- Comparison Between Spring and Fall Terms ---
plt.figure(figsize=(6,5))
sns.barplot(data=df, x="Term", y="Enrolled", ci=None)
plt.title("Enrollment Comparison: Spring vs Fall")
plt.ylabel("Average Enrolled Students")
plt.show()

# --- Department Enrollment Breakdown ---
dept_cols = ["Engineering Enrolled", "Business Enrolled", "Arts Enrolled", "Science Enrolled"]
dept_totals = df[dept_cols].sum().sort_values(ascending=False)

plt.figure(figsize=(7,5))
sns.barplot(x=dept_totals.index, y=dept_totals.values)
plt.title("Total Enrollment by Department (All Years)")
plt.ylabel("Total Students Enrolled")
plt.xticks(rotation=30)
plt.show()
```
